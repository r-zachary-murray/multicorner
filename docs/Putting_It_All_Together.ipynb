# Putting it all together

Finally, we can combine everything we've seen thus far and make a cornerplot for a single pre-prepared dataset. This is a good test of functionality for cases with many modes

```python
data = np.load('example_data.npz')['dataset']
```

```python
fig = mcorner(data,labels=['x','y'],confidence_ellipse=True,nsigma=5,inner_wspace=0.7,figsize=25,labelsize=14,ticksize=10,outer_wspace=0.1,outer_hspace=0.1)
fig.savefig('multicorner.png')
```
