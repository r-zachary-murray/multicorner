```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import corner
```

## Dataset One

```python
np.random.seed(1)

N, M, n_samples_per_mode, wide_range = 5, 3, 5000, 10

mu = np.random.uniform(-wide_range, wide_range, (N, M))
covariances = [np.dot(A := np.random.rand(M, M), A.T) for _ in range(N)]
all_samples1 = np.vstack([np.random.multivariate_normal(mu[i], covariances[i], n_samples_per_mode) for i in range(N)])

corner.corner(all_samples1, labels=[f"Dim {i+1}" for i in range(M)], show_titles=True)
plt.show()
```

```python
n_clusters = 9 # Choose the desired number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(all_samples1)
labels = kmeans.labels_ # Get cluster labels
```

```python
inds = np.random.randint(0,511,25)
```

```python
#this data will be used by tests
np.savez('../tests/data.npz',dataset=all_samples1,labels=labels,inds=inds,n=n_clusters)
```
